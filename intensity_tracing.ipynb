```python
def reflectivity(n1, n2, cos_theta):
    r0 = ((n1 / n2) - 1) / (n1 / n2 + 1)
    r0 = r0 * r0
    return r0 + (1 - r0)*((1-cos_theta) ** 5)
```

```python
n2 = 1.5
n1 = 1.0
```

```python
theta = np.linspace(0, np.pi, 50)
r = reflectivity(n1, n2, np.cos(theta))
plt.plot(np.cos(theta), 1 - r)
```
